```
human_graph = pd.read_table('/content/drive/My Drive/Bioinformatics/swc human scored_edges.txt', delim_whitespace=True, names=('protein', 'protein_connection', 'edge_weight'))
human_graph.head()
edges, human_col = human_graph.shape
print(edges)
```

```
def get_subgraph(protein):
  #print('inside get_subgraph: ', protein)
  sub_graph = human_graph.loc[human_graph['protein'] == str(protein)]
  #print(sub_graph)
  return sub_graph
```

```
# TEST
test_sub_graph = get_subgraph('A0AUZ9')
```

```
unique_proteins = human_graph.protein.unique()
convertion = {k: v for v, k in enumerate(unique_proteins)}
protein_length = len(convertion)

print(convertion)
def get_mapping(protein_value):
  if protein_value not in convertion:
    convertion[protein_value] = len(convertion)
    return convertion[protein_value]
  else:
    return convertion[protein_value]

```

```
# TEST
print(get_mapping('Q9UQ74'))
```

```
# MINIMAL CHANGE HERE
def appending(protein_subgraph):
  row, col = protein_subgraph.shape
  temp = ''
  for row_at in range(0, row):
    value = str(protein_subgraph.iat[row_at, 1])
    #print(value, ' ', 'converted: ', get_mapping(value))
    #temp = temp + str(get_mapping(value)) + ' ' + str( round( (protein_subgraph.iat[row_at, 2] * 10), 6) )+ ' '
    temp = temp + str(get_mapping(value)) + ' ' + '1'+ ' '
  #print(temp)
  return temp + '-1 -1'
```

```
# TEST
test_append = appending(test_sub_graph)
print(test_append)
```

```
formatted_human_data = open('formatted_human_data_for_one', 'w')
print('total length: ', len(unique_proteins))
counter = 1

first_line = str(protein_length) + ' ' + str(edges) + ' ' + '1'
formatted_human_data.write(first_line)
formatted_human_data.write('\n')
for protein in unique_proteins:
  if counter % 100 == 0:
    print('counter at: ', counter)
  counter+= 1

  protein_subgraph = get_subgraph(protein)
  adjacency = appending(protein_subgraph)
  #print(adjacency)
  formatted_human_data.write(adjacency)
  formatted_human_data.write('\n')
formatted_human_data.close()
print('done')
```
